## Detecting outliers

```python
max_threshold = data['price'].quantile(0.95)
max_threshold
```

```python
data[data.price > max_threshold]
```

```python
min_threshold = data['price'].quantile(0.05)
min_threshold
```

```python
data[data.price < min_threshold]
```

## Remove outliers 

```python
data[(data['price'] > min_threshold) | (data['price'] < max_threshold)]
```
