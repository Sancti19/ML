```python
col = ["battery_power","clock_speed","fc","int_memory","m_dep","mobile_wt","n_cores","pc","px_height","px_width","ram","sc_h","sc_w","talk_time"]

Q1 = train[col].quantile(0.25)
Q3 = train[col].quantile(0.75)
IQR = Q3-Q1
Range = IQR*1.5
new_train = train[~((train[col] < (Q1 - Range)) |  (train[col] > (Q3 + Range))).any(axis =1)]
```

## Creating X feature and Y target

```python
X_feature = new_train.drop('price_range', axis = 1)
Y_target = new_train['price_range']

display(X_feature.head())
display(Y_target.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_feature,Y_target, test_size = 0.25, random_state = 42)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## Model Building

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix
logreg = LogisticRegression()
logreg.fit(X_train,Y_train)

Y_pred = logreg.predict(X_test)

Accuract_Score = round(accuracy_score(Y_pred, Y_test)*100,3)
print(f'The accuracy score is {Accuract_Score}%')

```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k_range = range(2,25)

score = []

for k in k_range:
    knn  = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    score.append(accuracy_score(Y_pred,Y_test))
print(list(k_range))
print()
print(score)
print()
confusion_matrix = metrics.confusion_matrix(Y_test, Y_pred)
print(confusion_matrix)
```

- From  the list, n_neighbirs = 3 gave the highest accuracy score of 0.9292929292929293 

### Support Vector Machine

```python
from sklearn.svm import SVC

svm = SVC()

svm.fit(X_train,Y_train)

Y_pred = svm.predict(X_test)
Accuracy_score = (accuracy_score(Y_pred,Y_test))
print(f'The accuracy score is {Accuracy_score}')
print()
confusion_matrix = metrics.confusion_matrix(Y_test, Y_pred)
print(confusion_matrix)
```
